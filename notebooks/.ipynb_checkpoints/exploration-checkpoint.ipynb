# Data Preprocessing

```python
# Cleaning dataset from outlier
df_clean = df.copy()
# Masking person_age with age under 100 bcz age at 100+ up it's outlier
df_clean = df[df['person_age'] < 100]
# Limit it exp work at the limit 60
df_clean = df[df['person_emp_length'] <= 60]
# Handling missing values
# Fill loan_int_rate with mean 
df['loan_int_rate'].fillna(df['loan_int_rate'].mean(), inplace=True)
# Fill person_emp_length with median
df['person_emp_length'].fillna(df['person_emp_length'].median(), inplace=True)

# Check
print('Dataset After Cleaning')
print(f'Missing Values :\n{df.isna().sum()}')
df
```

```python
# Select features
num_feat = df.select_dtypes(include=[np.number]).drop('loan_status', axis=1).columns.tolist()
cat_feat = df.select_dtypes(include='object').columns.tolist()
targets = 'loan_status'

X = df[num_feat + cat_feat]
y = df[targets]

# Preprocessing data
preprocess = ColumnTransformer(transformers=[
        ('encoding', OneHotEncoder(drop='first'), cat_feat),
        ('scaling', StandardScaler(), num_feat)
    ], remainder='passthrough')

kf = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)

def splitting(X, y, random_state=42, test_size=0.2):
    '''This function is defined to split the 
    data into train and test.'''
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state, stratify=y)
    return X_train, X_test, y_train, y_test

X_train, X_test, y_train, y_test = splitting(X, y, random_state=42, test_size=0.2)
```

# Train model

```python
def train(X_train, y_train, preprocessor, cv):
    report = {}
    best_estimator = {}
    model = {
        'LogisticRegression': LogisticRegression(random_state=42, max_iter=1000),
        'XGBoost': XGBClassifier(random_state=42, n_jobs=-1),
        'LightGBM': LGBMClassifier(random_state=42, n_jobs=-1, verbose=-1)
    }
    params = {
        'LogisticRegression': {
            'model__C': [0.1, 1, 10],
            'model__penalty': ['l2'], 
            'model__solver': ['lbfgs', 'liblinear']
        },
        'XGBoost': {
            'model__learning_rate': [0.01, 0.1],
            'model__n_estimators': [100, 200],
            'model__max_depth': [3, 5, 7],
            'model__subsample': [0.8, 1.0], 
            'model__scale_pos_weight': [1, 3] 
        },
        'LightGBM': {
            'model__learning_rate': [0.01, 0.1],
            'model__n_estimators': [100, 200],
            'model__num_leaves': [31, 50], 
            'model__boosting_type': ['gbdt', 'dart'],
            'model__class_weight': [None, 'balanced'] 
        }
    }
    for model_name, model_init in model.items():
        pipeline = Pipeline([
            ('preprocess', preprocessor),
            ('smote', SMOTE(random_state=42)),
            ('model', model_init)
        ])
        model_params = params[model_name]
        # Grid search
        gs = GridSearchCV(pipeline,
                         param_grid=model_params,
                         cv=cv,
                         n_jobs=-1,
                         scoring='f1',
                         verbose=1)
        # Train
        gs.fit(X_train, y_train)

        # Keep best scores and best params
        best_f1 = gs.best_score_
        best_params = gs.best_params_

        print(f'{model_name} Best F1: {best_f1}')
        print(f'Params : {best_params}')

        report[model_name] = {
            'best_score':best_f1,
            'best_params':best_params
        }

        best_estimator[model_name] = gs.best_estimator_

    return report, best_estimator
```

```python
# Initialization with function train
reports, best_estimators = train(X_train, y_train, preprocess, kf)
```
